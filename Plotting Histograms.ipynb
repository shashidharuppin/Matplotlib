```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, rwidth=0.9)
```

```python
plt.hist(blood_sugar, bins = 4, color ='green')
```

```python
plt.hist(blood_sugar, bins = 4, color ='green',edgecolor ='black')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g',edgecolor ='black')
```

### Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], color = ['green','orange'],label = ['Men','Women'], edgecolor ='black')
plt.legend()
plt.show()
```

### histtype=step

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step',color='green')
```

### Horizontal Orientation

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='green',orientation = 'horizontal',edgecolor='black')
```
