```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Simple Bar Chart showing revenue of US Tech Companies`

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
```

```python
plt.bar(company,revenue, label = 'Revenue')
plt.ylabel('Revenue_in (Bln)')
plt.title ("Top US Companies Revenue")
plt.legend()
plt.show()
```

### Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue, label = 'Revenue')
plt.bar(company,profit, label = 'Profit')
plt.ylabel('Revenue_in (Bln)')
plt.title ("Top US Companies Revenue")
plt.legend()
plt.show()
```

```python
#Ploting the revenue and profit in the same bar graph side by side
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company) #Used to map the values to the string
plt.ylabel("Revenue_in (Bln)")
plt.title('Top US Companies Revenue')
plt.legend()
plt.show()
```

### Horizontal bar chart using barh function

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.xlabel("Revenue_in (Bln)")
plt.title('Top US Companies Revenue')
plt.legend()
plt.show()
```

```python
plt.barh(xpos,revenue, label="Revenue")
plt.barh(xpos,profit, label="Profit")

plt.yticks(xpos,company)
plt.xlabel("Revenue_in (Bln)")
plt.title('Top US Companies Revenue')
plt.legend()
plt.show()
```

```python
plt.barh(xpos-0.2,revenue,height =0.4, label="Revenue")
plt.barh(xpos+0.2,profit,height =0.4, label="Profit")

plt.yticks(xpos,company)
plt.xlabel("Revenue_in (Bln)")
plt.title('Top US Companies Revenue')
plt.legend()
plt.show()
```
