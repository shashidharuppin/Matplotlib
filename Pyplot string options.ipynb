```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day') #Labelling the X axis
plt.ylabel('Temperature') #Labelling the y axis
plt.title('Weather') # Labelling the title of the plot
plt.plot(x,y,'b+') # plotting x,y and b indicate color, + indicates marker
plt.show()
```

```python
plt.xlabel('Day') #Labelling the X axis
plt.ylabel('Temperature') #Labelling the y axis
plt.title('Weather') # Labelling the title of the plot
plt.plot(x,y,'g+--') # plotting x,y and g indicate color, + indicates marker and -- indicates spacing
plt.show()
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='blue',marker='+',linestyle='')
plt.show()
```

```python
# Same effect as 'g+--' format string
plt.plot(x,y,color='green',marker='+',linestyle='--')
plt.show()
```

### Marker Size 

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=10)
plt.show()
```

### Alpha property to control transparency of line chart

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
plt.show()
```

```python
plt.plot(x,y,'g<',alpha=0.7) # alpha can be specified on a scale 0 to 1
plt.show()
```

```python
plt.plot(x,y,'g<',alpha=0.2) # alpha can be specified on a scale 0 to 1
plt.show()
```
